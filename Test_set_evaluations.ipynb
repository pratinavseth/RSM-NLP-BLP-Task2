```python
import pandas as pd
import numpy as np
```

```python
!wget https://raw.githubusercontent.com/blp-workshop/blp_task2/main/data/test/blp23_sentiment_test_with_label.tsv
true_labels=pd.read_csv('blp23_sentiment_test_with_label.tsv', delimiter='\t')
```

```python
predict_results={'file1':'/kaggle/input/test-set-predictions/roberta.tsv', 'file2':'/kaggle/input/test-set-predictions/banglabertsmall.tsv', 'file3':'/kaggle/input/test-set-predictions/bengaliabusivemuril.tsv', 'file4':'/kaggle/input/test-set-predictions/sagorsarker.tsv', 'file5':'/kaggle/input/test-set-predictions/hatespeech.tsv', 'file6':'/kaggle/input/test-set-predictions/distilbert30.tsv', 'file7':'/kaggle/input/test-set-predictions/banglabert.tsv', 'file8':'/kaggle/input/test-set-predictions/banglishbert.tsv'}
```

```python
from sklearn.metrics import f1_score, accuracy_score, recall_score, precision_score
```

```python
f1score={}
y_true=true_labels['label']
for i in range(1,9):
    current_file='file'+str(i)
    df=pd.read_csv(predict_results[current_file], delimiter='\t')
    y_pred=df['label']
    f1s=f1_score(y_true, y_pred, average='micro')
    f1score[current_file]=f1s
f1score
```

```python
accuracy={}
y_true=true_labels['label']
for i in range(1,9):
    current_file='file'+str(i)
    df=pd.read_csv(predict_results[current_file], delimiter='\t')
    y_pred=df['label']
    acs=accuracy_score(y_true, y_pred)
    accuracy[current_file]=acs
accuracy
```

```python
recall={}
y_true=true_labels['label']
for i in range(1,9):
    current_file='file'+str(i)
    df=pd.read_csv(predict_results[current_file], delimiter='\t')
    y_pred=df['label']
    rs=recall_score(y_true, y_pred, average='weighted')
    recall[current_file]=rs
recall
```

```python
precision={}
y_true=true_labels['label']
for i in range(1,9):
    current_file='file'+str(i)
    df=pd.read_csv(predict_results[current_file], delimiter='\t')
    y_pred=df['label']
    ps=precision_score(y_true, y_pred, average='weighted')
    precision[current_file]=ps
precision
```
